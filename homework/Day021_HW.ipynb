### 1. 根據範例使用 selenuim 取出 Hahow 網站中所有課程的資料，並用一個適合的資料結構儲存（Hint: 將所有資料存在一個變數當中取出）

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./Data/chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html = browser.page_source
html[:500]
```

```python
soup = BeautifulSoup(html, "html5lib")
#soup
```

```python
course = []
course_title = soup.find_all("h4")
for title in course_title:
    course.append(title.text)

print(course)
```

```python
browser.close()
```

### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
browser = webdriver.Chrome(executable_path='./Data/chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html = browser.page_source
soup = BeautifulSoup(html, "html5lib")
#soup
```

```python
# 1. 先取出全部有幾頁
soup.find("div", class_="pagination-container").find_all("a")
```

```python
# 2. 每頁都取出資料放入 all_courses，pages為總頁數
pages = 30
all_courses = []

for i in range(pages):
    browser.get("https://hahow.in/courses?page={}".format(i+1))

    #取當前頁面
    html = browser.page_source
    soup = BeautifulSoup(html, "html5lib")

    #爬當前頁面的title
    course_title = soup.find_all("h4")
    for title in course_title:
        all_courses.append(title.text)
```

```python
browser.close()
print(all_courses)
```
