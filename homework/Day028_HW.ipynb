```python
# main.py

import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings


def main():
    #target_board = ['Gossiping', 'Stock']
    #target_board = ['Gossiping']
    target_board = ['Animation']
    process = CrawlerProcess(get_project_settings())
    for board in target_board:
        process.crawl('PTT', board=board)
        process.start()

if __name__ == '__main__':
    main()
```

```python
# 存好的JSON

import json
with open("./Data/PTT_crawler/crawled_data/.tmp.json.swp", encoding="utf-8") as j:
    data = json.load(j)
print(type(data))
print(type(data[0]))

for article in data:
    print("")
    print(article["url"])
    print(article['article_id'])
    print("")
    print("作者：", article['article_author'])
    print("標題：", article['article_title'])
    print("日期：", article['article_date'])
    print("")
    print(article['article_content'])
    print("IP：", article['ip'])
    print("總留言數:{}".format(article['message_count']['all']))
    print("推噓差:{}".format(article['message_count']['count']))
    print("推文數:{}".format(article['message_count']['push']))
    print("噓文數:{}".format(article['message_count']['boo']))
    print("箭頭數:{}".format(article['message_count']['neutral']))
    print("--------------------------------------------------------------------------------")
    for i in article['messages']:
        print(i['push_tag'], i['push_userid'], i['push_content'], i['push_ipdatetime'])
    print("================================================================================")
```
